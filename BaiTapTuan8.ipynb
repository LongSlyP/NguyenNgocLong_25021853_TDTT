```python
#27
def couple(ch1,ch2):
    return (ch1 == '(' and ch2 == ')') or (ch1 == '{' and ch2 == '}') or (ch1 == '[' or ch2 == ']') or (ch1 == '<' and ch2 == '>')
def check(text):
    lstopen = ['(','{','[','<']
    lstclose = [')','}',']','>']
    lst_word = []
    for x in text:
        if x in lstopen:
            lst_word.append(x)
        else:
            if len(lst_word) == 0:
                return False
            else:
                last = lst_word.pop()
                if not couple(last,x):
                    return False
    return len(lst_word) == 0
a = input()
print(check(a))
        
```

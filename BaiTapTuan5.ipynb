```python
def Max_Of_Two(a, b) -> int:
    if a > b:
        return a
    else:
        return b
while True:
    try:
        a = int(input("moi ban nhap vao so nguyen a: "))
        b = int(input("moi ban nhap vao so nguyen b: "))
        break
    except ValueError:
        print("nhap lai di")
    
print("so lon nhat la: ",Max_Of_Two(a, b))
```

```python
def swap(a,b) -> tuple[int, int]:
    a , b = b , a
    return a,b
while True:
    try:
        a = int(input("moi ban nhap vao so nguyen a: "))
        b = int(input("moi ban nhap vao so nguyen b: "))
        break
    except ValueError:
        print("nhap lai di")
a,b = swap(a,b)
print("gia tri cua a la: ",a)
print("gia tri cua b la: ",b)
```

```python
def kiemtra(n):
    
  
    if n < 2:
        return False
    elif n == 2: 
        return True
    else:
        for i in range(2,n):
            if n % i == 0:
                return False
            return True
try:
    n = int(input("moi ban nhap gia tri cua n: "))
except ValueError:
    print("day khong phai la so nguyen to")
if kiemtra(n):
    print("day la so nguyen to")
else:
    print("day khong phai la so nguyen to")
```

```python
import math
def sohoanhao(n):
    x = 0
    sqrt = math.isqrt(n)
    for i in range(1,sqrt+1):
        if n % i == 0:
            uoc1 = i
            uoc2 = n//i
            if uoc1 == uoc2:
                x += uoc1
            else:
                x += uoc1 + uoc2
    x = x - n
    if x == n:
        return True
    return False
while True:
    try:
        n = int(input("moi ban nhap vao so nguyen n: "))
        break
    except ValueError:
        print("thu lai di")
if n <= 0:
    print(f"{n} khong phai la so hoan hao")
else:
    if sohoanhao(n):
        print(f"{n} la so hoan hao")
    else:
        print(f"{n} khong la so hoan hao")
```

```python
def songuyen(i,k):
    x= 0
    
    for t in i:
        if k == t:
            return x
        x = x + 1
    return -1
i = list(map(int, input("moi ban nhap vao 1 day so nguyen: ").split()))
k = int(input())
print(songuyen(i,k))
    
```

```python
def giaithua(n):
    n1 = 1 
    for x in range(1,n+1):
        n1 *=  x
    return n1
while True:
    try:
        n = int(input("moi ban nhap vao so nguyen n:"))
        break
    except ValueError:
        print("nhap lai di")
if n == 0:
    print(f"{0}! = 1")
elif n < 0:
    print("so am khong co giai thua")
else:
    print(f"{n}! = {giaithua(n)}")
    
```

```python
def maytinh(a,b,operat):
    if operat == "*":
        ketqua = a*b
    elif operat == "+":
        ketqua = a+b
    elif operat == "-":
        ketqua = a-b,2
    elif operat == "/":
        if b == 0:
            print("loi phep tinh")
        else:
            ketqua = a/b
    return  f"{ketqua:.2f}"
a = float(input("moi ban nhap gia tri cua a: "))
b = float(input("moi ban nhap gia tri cua b: " ))
operat = input("moi nhap toan tu (* , / , + , -): ")
print("ket qua: ",maytinh(a,b,operat))
```

```python
def hamming(a,b):
    c = bin(a^b).count('1')
    return c
a = int(input())
b = int(input())
print(hamming(a,b))
```

```python
def tongn(n):
    x = 0
    while n != 0:
        x = x + n%10
        n //= 10
    return x
while True:
    try:
        n = int(input("moi ban nhap so nguyen duong n: "))
        if n <= 0:
            print("moi ban nhap lai")
            continue
        break
    except ValueError:
        print("nhap lai di")
print("tong cac chu so cua n la: ",tongn(n))
```

```python
def chu(a,b):
    if len(set(a)) == len(set(b)) == len(set(zip(a,b))) == len(set(b)):
        return True
    return False
a = str(input("moi nhap chuoi ky tu cua a:"))
b = str(input("moi nhap chuoi ky tu cua b: "))
print(chu(a,b))
    
```

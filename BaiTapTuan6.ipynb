```python
thuvien1 = {}
key1 = input("moi nhap vao cac key cho dict1: ").split()
while len(key1) != len(set(key1)):
    key1 = input("moi nhap vao lai cac key cho dict1(do bi trung key): ").split()
    
for i in key1:
    thuvien1[i] = int(input(f"moi nhap value cho  {i}:"))
thuvien2 = {}
key2 = input("moi nhap vao cac key cho dict2: ").split()
while len(key2) != len(set(key2)):
    key2 = input("moi nhap vao lai cac key cho dict1(do bi trung key): ").split()
thuvien3 = {}
for i in key2:
    thuvien2[i] = int(input(f"moi nhap value cho {i}: "))
for i in key1:
        thuvien3[i] = thuvien1[i]
for i in key2:
    if i not in thuvien3:
        thuvien3[i] = thuvien2[i]
    else:
        thuvien3[i] += thuvien2[i]
for key in sorted(thuvien3):
    print(key + ":" ,thuvien3[key])


        
```

```python
while True:
    try:
        songuyen = input("moi ban nhap vao list so nguyen(cach nhau boi khoang cach): ")
        chuoisonguyen = songuyen.split()
        number = [ int(x) for x in chuoisonguyen ]
        break
    except ValueError:
        print("ban nhap sai dinh dang roi")
k = int(input("moi nhap so nguyen duong k:"))
ketqua = []
for i in range(len(number)):
    for j in range(i+1,len(number)):
        if number[i] + number[j] == k:
            ketqua.append((number[i],number[j]))
ketqua.sort(key = lambda x : x[0])
print(ketqua)
           
```

```python
def loaibo(lst):
    result = []
    for i in lst:
        if i not in result:
            result.append(i)
    lst = list(result)
    return lst
danhsach1 = list(map(int, input("moi nhap vao danh sach cac so nguyen cua danhsach1: ").split()))
danhsach2 = list(map(int, input("moi nhap vao danh sach cac so nguyen cua danhsach2: ").split()))
danhsach1 = loaibo(danhsach1)
danhsach2 = loaibo(danhsach2)
phantulap = []
for i in danhsach1:
    for j in danhsach2:
        if i == j:
            phantulap.append(i)
print(phantulap)
```

```python
n,m,k = map(int, input("moi nhap 3 so n m k cach nhau boi dau cach: ").split())
mattrix = []
if k > m:
    print("khong ton tai thu k")
else:
    for i in range(n):
        while True:
            try:
                row = list(map(int , input(f"moi nhap vao phan tu cua hang {i + 1}: ").split()))
                if len(row) > m:
                    continue
                break
            except ValueError:
              print("phai nhap chu so")
        mattrix.append(row)
    x = len(mattrix)
    tong = 0
    for i in range(x):
        tong += mattrix[i][k-1]
    print(f"tong cac phan tu cua cot {k} la: {tong}")
    
    
        
    
```

```python
def somax(a,k):
    x = len(a)
    for i in range(x):
        if a[i] == k:
            return i
    return -1
a = list(map(int ,input("Moi nhap day so nguyen: ").split()))
k = int(input("moi nhap so nguyen duong k: "))
print(somax(a,k))
```

```python
def daytanggiam(x):
    y = len(x)
    for i in range(1,y-1):
        if x[i] > x[i-1] and x[i] > x[i+1] or x[i] < x[i-1] and x[i] < x[i+1]:
            return ("day vo danh")
        elif x[i] > x[i-1] and x[i] < x[i+1]:
            return ("day tang")
        else:
            return ("day giam")
x = list(map(int ,input("moi ban nhap day cac so nguyen: ").split()))
print(daytanggiam(x))
```

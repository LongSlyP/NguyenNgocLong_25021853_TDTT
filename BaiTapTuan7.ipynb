```python
def binary_sort(a,target):
    x = len(a)
    left = 0
    right = x -1
    while left <= right:
        mid = (left+right)//2
        if a[mid] == target:
            return mid 
        elif a[mid] < target:
            left = mid + 1
        elif a[mid] > target:
            right = mid - 1
    return -1
arr = list(map(int, input().split()))
target = int(input())
print(binary_sort(arr,target))
        
```

```python
def count_occurrences(arr, x) -> int:
    tong = 0
    for i in arr:
        if i == x:
            tong += 1
    return tong
arr = list(map(int ,input().split()))
x = int(input())
print(count_occurrences(arr, x))
             
```

```python
arr = list(map(int, input("moi nhap danh sach: ").split()))
counts = {}
for i in arr:
    counts[i] = counts.get(i,0) + 1
dct = {}
for i in counts:
    dct[i] = [i]*counts[i]
    if i + 1 in counts:
        dct[i].extend([i+1]*counts[i+1])
dodaimax = 0
for k,v in dct.items():
    if len(v) > dodaimax:
        dodaimax = len(v)
print(dodaimax)
    
```
